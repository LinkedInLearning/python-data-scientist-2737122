```python
mes = "Enero"
pais = "Espa√±a"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("ventas.csv")
```

```python
df_filtrado = df[(df['pais'] == pais) & (df['mes'] == mes)]
df_filtrado
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df_filtrado, x='producto', y='ventas')
plt.title("Ventas por producto en " + pais + " - " + mes)
plt.tight_layout()
plt.show()
```
